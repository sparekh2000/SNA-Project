```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('twitter.csv')
```

```python
df.drop([0],inplace=True)
```

```python
df.columns=['From','To']
df = df.astype(int) 
```

```python
graph = nx.from_pandas_edgelist(df,source="From",target="To",create_using=nx.DiGraph())
```

```python
degree_centrality = nx.degree_centrality(graph) 
degree_centrality
```

```python
eigenvector_centrality = nx.eigenvector_centrality(graph)
eigenvector_centrality
```

```python
katz_centrality = nx.katz_centrality_numpy(graph)
katz_centrality
```

```python
pr = nx.pagerank(graph, alpha = 0.8) 
pr 
```

```python
between_centrality=nx.betweenness_centrality(graph)
between_centrality
```

```python
close_centrality = nx.closeness_centrality(graph) 
close_centrality
```
