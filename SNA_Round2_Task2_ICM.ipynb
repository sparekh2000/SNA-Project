```python
import networkx as nx
import numpy as np,numpy.random
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G= nx.barabasi_albert_graph(20,4)
```

```python
nx.draw(G, with_labels=True,node_size=600,node_color="coral",edge_color="RebeccaPurple",node_shape="s",font_size=15)
```

```python

for iter in range(0,5):
  activated_nodes=[iter]
  current=[iter]
  print('Starting node is :',iter)
  time=-1

  while len(current)!=0:
    neighbors = [n for n in G.neighbors(current[0])]
    n=len(neighbors)
    degree=G.degree(current[0])
    # dirichlet generate n probabilities with sum=1
    probability=np.random.dirichlet(np.ones(degree))   
    flag=0
    for j in range(0,n):
      if (neighbors[j] not in activated_nodes):
        if flag==0:
          time=time+1
          flag=1
          # here we generate any random number between 0 and 1
        random_number=np.random.uniform(0,1) 
        
        # if random number > probability of that edge , then only targeted node can activate
        if (random_number>probability[j]) : 
        
          activated_nodes.append(neighbors[j])
          
          current.append(neighbors[j])
          
      
    current.pop(0)
  print('after time t=',time,'activated nodes are :',activated_nodes)
  print('Number of steps required :',time) 
  print("\n")     
  
```
